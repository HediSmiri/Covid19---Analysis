```python
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import accuracy_score
import numpy as np
import pandas as pd

data = pd.read_csv("covid_19_data.csv")

def clean_dataset(df):
    try:
        assert isinstance(df, pd.DataFrame), "df needs to be a pd.DataFrame"
        df.dropna(inplace=True)
        indices_to_keep = ~df.isin([np.nan, np.inf, -np.inf]).any(1)
        return df[indices_to_keep].astype(np.float64)
    except:
        pass

clean_dataset(data)
model = LinearRegression()
x = data.iloc[: ,  [5]].values
y = data.iloc[: , 6].values

X_train,X_test,y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=0)

# print(x.shape)
# print(y.shape)
model.fit(X_train,y_train)
y_pred= model.predict(X_test)


plt.scatter(x,y)
plt.plot(x, model.predict(x))
plt.xlabel("Daily New Cases")
plt.ylabel("Daily Deaths")
plt.title("New Cases vs New Deaths 04/14")
plt.show()

print(f"Linear Regression Accuracy Table : {model.score(X_test, y_test)}")
print("Absoulute Error : ",metrics.mean_absolute_error(y_test,y_pred))
print("Squared Error:",metrics.mean_squared_error(y_test,y_pred))
print("Root Squared Error:",np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
print(f'Model Coef : {model.coef_}')
```
